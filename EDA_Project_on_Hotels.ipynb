````python
number_of_reviews = []
average_rating = []

for item in overall_rating:
    if re.match(r'\(\d+ Ratings\)', item):
        number_of_reviews.append(int(re.findall(r'\d+', item)[0]))
    else:
        average_rating.append(item)

# Ensure both lists are of the same length by repeating elements as needed
min_len = min(len(number_of_reviews), len(average_rating))
number_of_reviews = number_of_reviews[:min_len]
average_rating = average_rating[:min_len]

df = pd.DataFrame({
    'number_of_reviews': number_of_reviews,
    'average_rating': average_rating
})

print(df)
````

````python
No_of_ppl_booked = []
for i in range(1,50):
    url = f"https://www.oyorooms.com/search?checkin=06%2F11%2F2024&checkout=07%2F11%2F2024&city=Bangalore&country=india&countryName=India&coupon=&filters%5Bcity_id%5D=4&guests=1&location=Bangalore%2C%20Karnataka%2C%20India&roomConfig%5B%5D=1&rooms=1&searchType=city&showSearchElements=false"
    page = requests.get(url,headers = request_header)
    soup = BeautifulSoup(page.text)
    for i in soup.find_all("div",class_="c-1sugwtq"):
        text=i.text
        if text:
           No_of_ppl_booked.append(text)
        else:
            No_of_ppl_booked.append(np.nan)
````

````python
len(No_of_ppl_booked)
````

````python
No_of_people_booked = []
Time_period = []

for item in No_of_ppl_booked:
    count_match = re.search(r'(\d+k?\+?) people', item)
    time_match = re.search(r'(today|last \d+ (hours|months))', item)
    
    if count_match and time_match:
        No_of_people_booked.append(count_match.group(1))
        Time_period.append(time_match.group(1))

# Create a DataFrame
df = pd.DataFrame({
    'No_of_people_booked': No_of_people_booked,
    'Time_period': Time_period
})

print(df)
````
